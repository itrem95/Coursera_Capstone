## 2. Read data from geojson file into a DataFrame

```python
def get_coordinates(data):
    if len(data['geometry']['coordinates'][0]) > 1:
        return data['geometry']['coordinates'][0][0]
    
    return data['geometry']['coordinates'][0][0][0]
```

```python
with open('mo.geojson') as json_data:
    mo_data = json.load(json_data)
```

```python
# define the dataframe columns
column_names = ['Neighbourhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
moscow_districts_df = pd.DataFrame(columns=column_names)
```

```python
for data in mo_data['features']:
    district_name = data['properties']['NAME'].strip()
    longitude, latitude = get_coordinates(data)
    
    moscow_districts_df = moscow_districts_df.append({'Neighbourhood': district_name,
                                          'Latitude': latitude,
                                          'Longitude': longitude}, ignore_index=True)
```

```python
moscow_districts_df
```

```python
# save the DataFrame as CSV file
moscow_districts_df.to_csv("moscow_districts_df.csv", index=False)
```

## 3. Create a map of Moscow with boroughs superimposed on top

```python
# get the coordinates of Moscow
address = 'Moscow, Russia'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Moscow are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Moscow using latitude and longitude values
map_moscow = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, neighbourhood in zip(moscow_districts_df['Latitude'], moscow_districts_df['Longitude'],
                                  moscow_districts_df['Neighbourhood']):
    label = '{}'.format(neighbourhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_moscow)  
```

```python
# save the map as HTML file
map_moscow.save('map_moscow.html')
```

## 4. Use the Foursquare API to explore the neighborhoods

```python
CLIENT_ID = '' # your Foursquare ID
CLIENT_SECRET = '' # your Foursquare Secret
VERSION = '20180604'

print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

**Now, let's get the top 100 venues that are within a radius of 2000 meters.**

```python
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(moscow_districts_df['Latitude'], moscow_districts_df['Longitude'], 
                                   moscow_districts_df['Neighbourhood']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['Neighborhood', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df.head()
```

## 5. Analyze Each Neighborhood

```python
# one hot encoding
moscow_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
moscow_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [moscow_onehot.columns[-1]] + list(moscow_onehot.columns[:-1])
moscow_onehot = moscow_onehot[fixed_columns]

print(moscow_onehot.shape)
moscow_onehot.head()
```

**Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category**

```python
moscow_onehot_grouped = moscow_onehot.groupby(['Neighborhoods']).mean().reset_index()

print(moscow_onehot_grouped.shape)
moscow_onehot_grouped
```

```python
len(moscow_onehot_grouped[moscow_onehot_grouped['Gym / Fitness Center'] > 0])
```

**Create a new DataFrame for Gym / Fitness Center data only**

```python
moscow_gyms = moscow_onehot_grouped[['Neighborhoods','Gym / Fitness Center']]
```

```python
moscow_gyms.head()
```

## 6. Cluster Neighbourhoods

Run k-means to cluster the neighborhoods in Moscow into 3 clusters.

```python
k_clusters = 3

moscow_clustering = moscow_gyms.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=k_clusters, random_state=0).fit(moscow_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
moscow_merged = moscow_gyms.copy()

# add clustering labels
moscow_merged["Cluster Labels"] = kmeans.labels_
```

```python
moscow_merged.rename(columns={"Neighborhoods": "Neighbourhood"}, inplace=True)
moscow_merged.head()
```

```python
moscow_merged = moscow_merged.join(moscow_districts_df.set_index("Neighbourhood"), on="Neighbourhood")
```

```python
moscow_merged.head()
```

```python
moscow_merged.sort_values(["Cluster Labels"], inplace=True)
moscow_merged
```

**Let's visualize the resulting clusters**

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k_clusters)
ys = [i+x+(i*x)**2 for i in range(k_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neigh, cluster in zip(moscow_merged['Latitude'], moscow_merged['Longitude'],
                                  moscow_merged['Neighbourhood'], moscow_merged['Cluster Labels']):
    label = folium.Popup(str(neigh) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
```

```python
# save the map as HTML file
map_clusters.save('map_clusters.html')
```
